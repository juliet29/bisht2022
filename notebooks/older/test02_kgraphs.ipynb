```python
import networkx as nx
import random 
```

```python
def plot_graph(G, pos = None):
    labels = {node: node for node in G.nodes}

    if not pos:
        pos = nx.planar_layout(G)
    nx.draw_networkx_nodes(G, pos)
    nx.draw_networkx_edges(G,pos )
    nx.draw_networkx_labels(G, pos, labels=labels);

    return G, pos
```

```python
K1 = nx.random_regular_graph(1, 2)
plot_graph(K1)
```

```python
node_connectivity = nx.node_connectivity(K1)
print("Node Connectivity:", node_connectivity)
```

```python
K2 = nx.random_regular_graph(2, 3)
plot_graph(K2)
```

```python
node_connectivity = nx.node_connectivity(K2)
print("Node Connectivity:", node_connectivity)
```

```python
plot_graph(nx.random_regular_graph(d=2, n=3))
```

```python
def n_connected_graph(connectivity=2, num_nodes=3):
    # Define the number of nodes in the graph

    # Create an empty graph
    G = nx.Graph()

    # Add nodes to the graph
    G.add_nodes_from(range(num_nodes))

    # Iterate over each node and add edges until each node has at least x  connections
    for node in G.nodes():
        while G.degree[node] < connectivity:
            # Choose a random node to connect to
            target = random.choice(list(G.nodes()))
            # Make sure the chosen node is not the same as the current node and they are not already connected
            if target != node and not G.has_edge(node, target):
                G.add_edge(node, target)
    # TODO: add time out / checks 

    # Print the edges of the generated graph
    print("Edges of the graph:")
    print(G.edges())

    return G
```

```python
Gtri, _ = plot_graph(n_connected_graph())
```

```python
Gtri2, _ = plot_graph(n_connected_graph(1,3))
```

```python
Gtri3, _ = plot_graph(n_connected_graph(connectivity=3,num_nodes=4))
```
