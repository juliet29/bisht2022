```python
def plot_planar_embed(embed: nx.PlanarEmbedding):
    pos = nx.combinatorial_embedding_to_pos(embed)
    nx.draw_networkx(nx.Graph(embed), pos)

def plot_just_planar(G: nx.Graph):
    pos = nx.planar_layout(G)
    nx.draw_networkx(G, pos)


```

```python
# Define the combinatorial embedding
embedding = {
    1: [2, 3, 4],
    2: [1, 4, 5],
    3: [1, 4, 6],
    4: [1, 2, 3, 5, 6],
    5: [2, 4, 6],
    6: [3, 4, 5]
}

# Create a NetworkX graph
G = nx.Graph()

# Add nodes
G.add_nodes_from(embedding.keys())

# Add edges based on the embedding
for v, neighbors in embedding.items():
    for u in neighbors:
        if (u, v) not in G.edges() and (v, u) not in G.edges():
            G.add_edge(u, v)

# Define positions based on the embedding (manually or using planar_layout)
pos = {
    1: (0, 0),
    2: (1, 0),
    3: (0, 1),
    4: (0.5, 0.5),
    5: (1, 1),
    6: (0.5, 1.5)
}
```

```python
nx.draw_networkx(G, pos)
```
