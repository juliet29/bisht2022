```python
def seperating_triangle_check(G):
    l3_cycles = sorted(nx.simple_cycles(G, 3))
    m = len(list(G.edges))
    n = len(list(G.nodes))
    ic(len(l3_cycles), m, n, m-n+1);
    assert len(l3_cycles) == m-n+1
    
```

```python
seperating_triangle_check(G_comb_tri)
```
